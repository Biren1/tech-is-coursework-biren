```python
# import datetime module
import datetime

# .now() is a built-in datetime function
time_now = datetime.datetime.now()

# print time now
print(time_now)

# print year only
print(time_now.year)

# print day only
print(time_now.strftime("%A"))
```

```python
new_date = datetime.datetime(2020, 8, 1)

print(new_date)
```

```python

# Create a Data
x = datetime.datetime(2020, 8, 1)

# Get the Weekday name
print(x.strftime("%A"))

# Get the Month name
print(x.strftime("%B"))

# Get the Weekday name
print(x.strftime("%d"))

# Get the Weekday name
print(x.strftime("%Y"))
```

```python
sq_root = math.sqrt(90) # finds the square root

print(sq_root)
```
