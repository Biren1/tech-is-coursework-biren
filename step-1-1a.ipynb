```

# a and b are true
a = True 
b = True

print(a and b)

# only a is true 
a = True 
b = False

print(a and b)

# only b is true 
a = False 
b = True

print(a and b)

# a and b are both false
a = False 
b = False

print(a and b)
```

```

## 'or' operator 

# a and b are true
a = True 
b = True

print(a or b)

# only a is true 
a = True 
b = False

print(a or b)

# only b is true 
a = False 
b = True

print(a or b)

# a and b are both false
a = False 
b = False

print(a or b)
```
